```python
import os
import sys
import pandas as pd
import numpy as np
from dotenv import load_dotenv
from datetime import datetime

sys.path.insert(0, os.path.abspath('../src'))
from db.connection import read_sql

load_dotenv()
print("✓ Imports OK")
```

```python
query_sys_peak = """
SELECT 
    TIMESTAMP, LOG_TYPE, HTTP_STATUS_CODE, CLIENT_IP, SERVICE_ID
FROM SAP_SYSTEM_LOGS
WHERE EXTRACT(HOUR FROM TIMESTAMP) BETWEEN 8 AND 18
    AND DAYOFWEEK(TIMESTAMP) BETWEEN 2 AND 6
LIMIT 100000
"""
df_sys_peak = read_sql(query_sys_peak)
print(f"System PEAK: {len(df_sys_peak)} records")
print(df_sys_peak.head())
```

```python
query_sys_offpeak = """
SELECT 
    TIMESTAMP, LOG_TYPE, HTTP_STATUS_CODE, CLIENT_IP, SERVICE_ID
FROM SAP_SYSTEM_LOGS
WHERE (EXTRACT(HOUR FROM TIMESTAMP) < 8 OR EXTRACT(HOUR FROM TIMESTAMP) > 18)
    OR DAYOFWEEK(TIMESTAMP) NOT BETWEEN 2 AND 6
LIMIT 100000
"""
df_sys_offpeak = read_sql(query_sys_offpeak)
print(f"System OFFPEAK: {len(df_sys_offpeak)} records")
print(df_sys_offpeak.head())
```

```python
query_llm_peak = """
SELECT 
    TIMESTAMP, LOG_TYPE, LLM_MODEL_ID, LLM_STATUS, LLM_COST_USD, LLM_RESPONSE_TIME_MS
FROM SAP_LLM_LOGS
WHERE EXTRACT(HOUR FROM TIMESTAMP) BETWEEN 8 AND 18
    AND DAYOFWEEK(TIMESTAMP) BETWEEN 2 AND 6
LIMIT 100000
"""
df_llm_peak = read_sql(query_llm_peak)
print(f"LLM PEAK: {len(df_llm_peak)} records")
print(df_llm_peak.head())
```

```python
query_llm_offpeak = """
SELECT 
    TIMESTAMP, LOG_TYPE, LLM_MODEL_ID, LLM_STATUS, LLM_COST_USD, LLM_RESPONSE_TIME_MS
FROM SAP_LLM_LOGS
WHERE (EXTRACT(HOUR FROM TIMESTAMP) < 8 OR EXTRACT(HOUR FROM TIMESTAMP) > 18)
    OR DAYOFWEEK(TIMESTAMP) NOT BETWEEN 2 AND 6
LIMIT 100000
"""
df_llm_offpeak = read_sql(query_llm_offpeak)
print(f"LLM OFFPEAK: {len(df_llm_offpeak)} records")
print(df_llm_offpeak.head())
```
